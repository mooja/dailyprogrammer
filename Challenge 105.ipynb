```python
def find_anagrams(word, dictionary):
    anagrams = []
    for dictword in dictionary:
        if len(dictword) == len(word):
            if dictword != word and sorted(dictword) == sorted(word):
                anagrams.append(dictword)
    return anagrams


test_dict = ['a', 'dog', 'god', 'sausage']
find_anagrams("gdo", test_dict)
```

```python
def unscramble_words(text, dictionary):
    anagr_dict = {w.lower(): find_anagrams(w, dictionary) 
                    for w in text.strip().split()}
    
    # filter out empty anagram lists
    anagr_dict = {k: v for k, v in anagr_dict.iteritems() if v}
    return anagr_dict


test_dict = ['a', 'dog', 'god', 'sausage']
test_text = "god i hate this sausaeg"
unscramble_words(test_text, test_dict)
```
